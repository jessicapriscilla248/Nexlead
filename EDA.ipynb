## Relationship Between Features

```python
# Cek relationship antara Department dan Previous Year Rating
plt.figure(figsize=(15, 7))
plt.title('Hubungan antara department dengan previous year rating')
sns.boxplot(data=data, x=data["department"], y=data["previous_year_rating"])
plt.xlabel("Department")
plt.ylabel("Previous Year Rating")
plt.show()
```

Rating terbaik didapatkan di department R&D, HR, dan Operations

```python
plt.figure(figsize=(15, 7))
plt.title('Hubungan antara Number of Training dengan Previous Year Rating')
sns.lineplot(data=data, x=data["no_of_trainings"], y=data["previous_year_rating"], color="red", ci=None, marker='o')
plt.xlabel("Number of Training")
plt.ylabel("Previous Year Rating")
plt.show()
```

Rating paling tinggi saat pegawai mengikuti 8 kali training, yaitu 4.5. Sehingga dapat disimpulkan, bahwa:
1. 1-3 kali training, maka pegawai hanya bisa mendapatkan rating rata-rata karena skillnya mungkin masih standard karena baru sekali.
2. 7 kali training menghasilkan rating yang sangat rendah.

Dapat dilihat bahwa grafik naik turun dan bisa disimpulkan, bahwa jumlahnya training tidak dapat dijadikan patokan rating akan baik.

```python

plt.title('Hubungan antara Length of Service dengan Previous Year Rating')
sns.lineplot(data=data, x=data["length_of_service"], y=data["previous_year_rating"], color="red", ci=None, marker='o')
plt.xlabel("Length of Service")
plt.ylabel("Previous Year Rating")
plt.show()
```

```python

plt.title('Hubungan antara Age dengan Length of Service')
sns.lineplot(data=data, x=data["length_of_service"], y=data["age"], color="red", ci=None, marker='o')
plt.xlabel("Length of Service")
plt.ylabel("Age")
plt.show()
```

```python

plt.title('Hubungan antara KPI More than 80% dengan Previous Year Rating')
sns.boxplot(data=data, x=data["KPIs_met_more_than_80"], y=data["previous_year_rating"])
plt.xlabel("KPI More than 80%")
plt.ylabel("Previous Year Rating")
plt.show()
```

```python

plt.title('Hubungan antara Awards Won dengan Previous Year Rating')
sns.boxplot(data=data, x=data["awards_won"], y=data["previous_year_rating"])
plt.xlabel("Awards Won")
plt.ylabel("Previous Year Rating")
plt.show()
```

- Median rating pegawai yang menang penghargaan lebih tinggi daripada yang tidak.
- Karyawan yang menang penghargaan kebanyakan memiliki rating tinggi.
- Ada sedikit outlier di kelompok yang tidak menang penghargaan yang memiliki rating rendah.

Maka dapat disimpulkan bahwa 
- Pegawai dengan rating tinggi bisa berkesempatan mendapatkan penghargaan.
- Penghargaan bisa saja diberikan berdasarkan rating di tahun sebelumnya.

```python
# Cek relationship antara Department dan Average Training Score
plt.figure(figsize=(15, 7))
plt.title('Hubungan antara department dengan Average Training Score')
sns.boxplot(data=data, x=data["department"], y=data["avg_training_score"])
plt.xlabel("Department")
plt.ylabel("Average Training Rating")
plt.show()
```

- Departemen HR, Sales & Marketing, dan Finance memiliki rating dengan rata-ratanya rendah dan memiliki banyak outlier (rating 50-60).
- Departemen dengan rating pelatihan tinggi dan konsisten (R&D dan Analytics) bisa dijadikan model untuk program pelatihan di departemen lain (rating 8085).
- Departemen dengan banyak outlier (Operations dan Legal) bisa diteliti untuk diperbaiki mengapa terdapat perbedaan besar antar pegawai.

```python
# Cek relationship antara Region dan Previous Year Rating
plt.figure(figsize=(20, 7))
plt.title('Hubungan antara Region dengan Previous Year Rating')
sns.boxplot(data=data, y=data["previous_year_rating"], x=data["region"])
plt.xlabel("Region")
plt.ylabel("Previous Year Training")
plt.xticks(rotation=45, ha='right')
plt.show()
```

Region tidak mempengaruhi rating tahun sebelumnya

```python
# Cek relationship antara Region dan Average Training Score
plt.figure(figsize=(20, 7))
plt.title('Hubungan antara Region dengan Average Training Score')
sns.boxplot(data=data, y=data["avg_training_score"], x=data["region"])
plt.xlabel("Region")
plt.ylabel("Average Training Score")
plt.xticks(rotation=45, ha='right')
plt.show()
```

Region_7, region_29, region_28, region_33 yang memiliki skor rata-rata pelatihan lebih tinggi.
